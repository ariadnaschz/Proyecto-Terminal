Aqui puedes visualizar como es la dinamica de propagación en distintos tipos de familias de graficas:

Variables 
* n : número de vértices para la gráfica  
* m : número de vértices solo para las Rejillas y Bipartitas Completas 
* G : tipo de familia puede ser StarGraph(n), CompleteGraph(n), CycleGraph(n), PathGraph(n), WheelGraph(n), CompleteBipartiteGraph(n,m) y Rejilla(n,m)
* R : el número mínimo de vecinos infectados necesarios para que un vértice muerto se regenere como infectada.
* estados :  puedes llamas a la función [EstadoInicial(G)] para que te de un estado inicial aleatorio, o puedes ponerlo manualmente dependiendo del totala de vértices que elejiste por ejemplo [[1, 1, 1, 0, 0, 0, 0, 0]] 

```python
#Ejemplo
n=2
m=3 
G=Rejillas(n,m)
R=2
estados=[EstadoInicial(G)]
dibujos=[G.plot(vertex_colors=HacerDicCol(estados))]
for i in srange(10):
    estados.append(Paso(G, estados[i], R))
    dibujos.append(G.plot(vertex_colors=HacerDicCol(estados[i])))
print(estados[i])
gifi=animate(dibujos)
gifi.show()
```
